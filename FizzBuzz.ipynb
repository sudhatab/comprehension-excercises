```python
for fizbuz in range(101):
    if fizbuz % 3 == 0 and fizbuz % 5 == 0:
        print("fizzbuzz")

    elif fizbuz % 3 == 0:
        print("fizz")
        
    elif fizbuz % 5 == 0:
        print("buzz")
        
    print(fizbuz)
```

```python
fizbuz = ["fizzbuzz" if (fizbuz % 15 == 0) else ("fizz" if (fizbuz % 3 == 0) else "buzz" if (fizbuz % 5 == 0) else fizbuz) for fizbuz in range(101)]
print(fizbuz)
```
